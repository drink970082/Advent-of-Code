```python
def read(file):
    lock, key = [], []
    with open(file, "r") as f:
        while top_row := f.readline().strip():
            schematic = [list(top_row)]
            while line := f.readline().strip():
                schematic.append(list(line))
            height = []
            for col in range(len(schematic[0])):
                height.append(sum(1 for row in schematic if row[col] == "#") - 1)
            if top_row == "#####":
                lock.append(height)
            else:
                key.append(height)
    return lock, key
```

```python
from collections import defaultdict

key_lock_pair = defaultdict(list)
lock, key = read("input.txt")
for l in lock:
    for h in range(len(l)):
        key_lock_pair[tuple(l)].append(set())
        for i, k in enumerate(key):
            if l[h] + k[h] <= 5:
                key_lock_pair[tuple(l)][h].add(i)
unique_pairs = 0             
for k, v in key_lock_pair.items():
    pair = v[0]
    for i in range(1, len(v)):
        pair = pair.intersection(v[i])
    unique_pairs += len(pair)
print(unique_pairs)
```
